```python
import pandas as pd
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
table_names_query= "SELECT name FROM sqlite_master WHERE type='table'"
conn = sqlite3.connect('data/movies.db')
```

```python
conn.execute(table_names_query).fetchall()
```

```python
titles=pd.read_sql('''

SELECT * 
FROM imdbTitleBasics

''', conn         

           )

titles
```

```python
ratings = pd.read_sql('''

SELECT * 
FROM imdbTitleRatings

''', conn         

           )

ratings
```

```python
movie_gross = pd.read_sql('''

SELECT * 
FROM bomMoviegross

''', conn         

           )

movie_gross
```

```python
big_movie_gross=movie_gross[movie_gross['domestic_gross']>=100000000]
```

```python
big_movie_gross
```

```python
big_movie_gross=big_movie_gross.drop(columns='foreign_gross', axis=1)
```

```python
big_movie_gross
```

```python
fig, ax = plt.subplots(1,1, figsize=(8,7))

x = big_movie_gross['year']
plt.style.use('ggplot')
plt.hist(x, bins = 25)
plt.show
ax.set_title('Movies Where Gross > One Hundred Million', fontsize=18)
ax.set_ylabel('Number of Movies', fontsize=18)
ax.set_xlabel('Year of Release Date', fontsize=18)
```

```python
fig, ax = plt.subplots(1,1, figsize=(8,7))

x = big_movie_gross['studio']
plt.style.use('ggplot')
plt.hist(x, bins = 50)
plt.show
ax.set_title('Movies Where Gross > One Hundred Million', fontsize=18)
ax.set_ylabel('Number of Movies', fontsize=18)
ax.set_xlabel('Studio', fontsize=18)
```
